## 2-1. 도보 시간/거리 계산하기 (SK open API 사용)

```python
import requests
import json

def get_tmap_pedestrian_route(startX, startY, endX, endY, app_key):
    url = "https://apis.openapi.sk.com/tmap/routes/pedestrian"

    headers = {
        "Content-Type": "application/json",
        "appKey": app_key
    }

    data = {
        "startX": startX,
        "startY": startY,
        "endX": endX,
        "endY": endY,
        "reqCoordType": "WGS84GEO",
        "resCoordType": "WGS84GEO",
        "startName": "출발지",
        "endName": "도착지"
    }

    response = requests.post(url, headers=headers, data=json.dumps(data))
    
    print("📡 응답 코드:", response.status_code)  # 200이면 정상
    try:
        result = response.json()
        print("📦 응답 결과:", json.dumps(result, indent=2, ensure_ascii=False))
    except Exception as e:
        print("❗ JSON 변환 실패:", e)
        print("응답 원문:", response.text)
        return None, None

    if 'features' in result and result['features']:
        total_distance = result['features'][0]['properties'].get('totalDistance')
        total_time = result['features'][0]['properties'].get('totalTime')
        return total_distance, total_time
    else:
        print("❗ 'features' 키가 없거나 비어 있음")
        return None, None


# 테스트
app_key = "KXgnElYbnf9P4rPyZzanN91cHMyScabF1ZY4ifZR"
startX = 127.052427705027
startY = 37.5125058180695
endX = 127.098937173285
endY = 37.4044515123315

distance, time = get_tmap_pedestrian_route(startX, startY, endX, endY, app_key)

print("🛣️ 거리(m):", distance)
print("⏱️ 시간(초):", time)

```

## 2-2. 자동차 시간/거리 계산하기 (SK open API 사용)

```python
import requests

def get_car_route_time_distance(startX, startY, endX, endY, appKey):
    url = f"https://apis.openapi.sk.com/tmap/routes?version=1&format=json&appKey={appKey}"
    
    headers = {
        "Content-Type": "application/json"
    }
    
    body = {
        "startX": str(startX),
        "startY": str(startY),
        "endX": str(endX),
        "endY": str(endY),
        "reqCoordType": "WGS84GEO",
        "resCoordType": "WGS84GEO",
        "startName": "출발지",
        "endName": "도착지",
        "searchOption": 0,
        "trafficInfo": "Y"  # 불리언 대신 "Y" 문자열로 변경
    }
    
    response = requests.post(url, headers=headers, json=body)
    
    if response.status_code == 200:
        result = response.json()
        summary = result.get("features")[0].get("properties")
        distance_m = summary.get("totalDistance")
        time_s = summary.get("totalTime")
        
        print(f"거리: {distance_m} m")
        print(f"시간: {time_s} 초")
        
        return distance_m, time_s
    else:
        print(f"API 요청 실패: {response.status_code}")
        print(response.text)
        return None, None


if __name__ == "__main__":
    startX = 127.052427705027
    startY = 37.5125058180695
    endX = 127.098937173285
    endY = 37.4044515123315
    appKey = "KXgnElYbnf9P4rPyZzanN91cHMyScabF1ZY4ifZR"
    
    get_car_route_time_distance(startX, startY, endX, endY, appKey)


```

## 2-3-1. 대중교통 시간/거리 계산하기 (SK open API 사용)

```python
import requests

def get_transit_route(startX, startY, endX, endY, appKey):
    url = f"https://apis.openapi.sk.com/transit/routes?lang=0&format=json&appKey={appKey}"
    
    headers = {
        "Content-Type": "application/json"
    }

    body = {
        "startX": str(startX),
        "startY": str(startY),
        "endX": str(endX),
        "endY": str(endY),
        "count": 1  # 최적 경로 1개만 요청
    }

    response = requests.post(url, headers=headers, json=body)

    if response.status_code == 200:
        data = response.json()
        try:
            path = data["metaData"]["plan"]["itineraries"][0]
            totalTime = path["totalTime"]  # 초 단위
            totalDistance = path["totalDistance"]  # 미터 단위

            print("🚇 대중교통 경로 탐색 결과")
            print(f"🕒 소요 시간: {totalTime // 60}분 {totalTime % 60}초")
            print(f"📏 이동 거리: {totalDistance} m")

            return totalDistance, totalTime
        except (KeyError, IndexError):
            print("❗ 경로를 찾을 수 없습니다. 좌표를 다시 확인해 주세요.")
            return None, None
    else:
        print(f"❌ API 요청 실패: {response.status_code}")
        print(response.text)
        return None, None


# 여기에 입력값만 바꾸세요
if __name__ == "__main__":
    # ✅ 출발지 경도, 위도 (예: 위랩스)
    startX = 127.052427705027
    startY = 37.5125058180695

    # ✅ 도착지 경도, 위도 (예: SK쉴더스)
    endX = 127.098937173285
    endY = 37.4044515123315

    # ✅ 본인의 Tmap 발급 앱키
    appKey = "KXgnElYbnf9P4rPyZzanN91cHMyScabF1ZY4ifZR" 

    # 실행
    get_transit_route(startX, startY, endX, endY, appKey)

```

## 2-3-2. 대중교통 시간/거리 계산하기(구간별 교통수단(지하철/버스/도보)까지 출력) (선택)

```python
import requests

def get_transit_route(startX, startY, endX, endY, appKey):
    url = f"https://apis.openapi.sk.com/transit/routes?lang=0&format=json&appKey={appKey}"
    
    headers = {
        "Content-Type": "application/json"
    }

    body = {
        "startX": str(startX),
        "startY": str(startY),
        "endX": str(endX),
        "endY": str(endY),
        "count": 1
    }

    response = requests.post(url, headers=headers, json=body)

    if response.status_code == 200:
        data = response.json()
        try:
            path = data["metaData"]["plan"]["itineraries"][0]
            totalTime = path["totalTime"]
            totalDistance = path["totalDistance"]

            print("🚇 대중교통 경로 탐색 결과")
            print(f"🕒 총 소요 시간: {totalTime // 60}분 {totalTime % 60}초")
            print(f"📏 총 이동 거리: {totalDistance} m\n")

            print("📍 경로 상세:")
            for idx, leg in enumerate(path["legs"]):
                section_type = leg["mode"]  # WALK, BUS, SUBWAY
                distance = leg["distance"]
                section_time = leg["sectionTime"]

                if section_type == "WALK":
                    print(f"  {idx+1}. 🚶 도보 이동 - 거리: {distance}m, 시간: {section_time}초")
                elif section_type == "BUS":
                    bus_line = leg["route"]  # 버스 번호
                    start = leg["start"]["name"]
                    end = leg["end"]["name"]
                    print(f"  {idx+1}. 🚌 버스 {bus_line} → {start} → {end} ({distance}m, {section_time}초)")
                elif section_type == "SUBWAY":
                    subway_line = leg["route"]  # 지하철 노선명
                    start = leg["start"]["name"]
                    end = leg["end"]["name"]
                    print(f"  {idx+1}. 🚇 지하철 {subway_line} → {start} → {end} ({distance}m, {section_time}초)")
                else:
                    print(f"  {idx+1}. ❓ 기타 수단 ({section_type}) - 거리: {distance}m")

            return totalDistance, totalTime
        except (KeyError, IndexError):
            print("❗ 경로를 찾을 수 없습니다. 좌표를 다시 확인해 주세요.")
            return None, None
    else:
        print(f"❌ API 요청 실패: {response.status_code}")
        print(response.text)
        return None, None


# 여기에 입력값만 바꾸세요
if __name__ == "__main__":
    # ✅ 출발지 경도, 위도 (예: 위랩스)
    startX = 127.052427705027
    startY = 37.5125058180695

    # ✅ 도착지 경도, 위도 (예: SK쉴더스)
    endX = 127.098937173285
    endY = 37.4044515123315

    # ✅ 본인의 Tmap 발급 앱키
    appKey = "KXgnElYbnf9P4rPyZzanN91cHMyScabF1ZY4ifZR"

    # 실행
    get_transit_route(startX, startY, endX, endY, appKey)

```

## 3. 전체 코드 (세 방식 통합 + 최적 경로 판단)

```python
import requests

def get_walk_route(startX, startY, endX, endY, appKey):
    url = f"https://apis.openapi.sk.com/tmap/routes/pedestrian?version=1&format=json&appKey={appKey}"
    headers = {"Content-Type": "application/json"}
    body = {
        "startX": str(startX),
        "startY": str(startY),
        "endX": str(endX),
        "endY": str(endY),
        "reqCoordType": "WGS84GEO",
        "resCoordType": "WGS84GEO",
        "startName": "출발지",
        "endName": "도착지"
    }
    response = requests.post(url, headers=headers, json=body)
    if response.status_code == 200:
        try:
            props = response.json()["features"][0]["properties"]
            return ("도보", props["totalDistance"], props["totalTime"])
        except:
            return ("도보", None, None)
    return ("도보", None, None)

def get_car_route(startX, startY, endX, endY, appKey):
    url = f"https://apis.openapi.sk.com/tmap/routes?version=1&format=json&appKey={appKey}"
    headers = {"Content-Type": "application/json"}
    body = {
        "startX": str(startX),
        "startY": str(startY),
        "endX": str(endX),
        "endY": str(endY),
        "reqCoordType": "WGS84GEO",
        "resCoordType": "WGS84GEO",
        "startName": "출발지",
        "endName": "도착지",
        "searchOption": 0,
        "trafficInfo": "Y"
    }
    response = requests.post(url, headers=headers, json=body)
    if response.status_code == 200:
        try:
            props = response.json()["features"][0]["properties"]
            return ("자동차", props["totalDistance"], props["totalTime"])
        except:
            return ("자동차", None, None)
    return ("자동차", None, None)

def get_transit_route(startX, startY, endX, endY, appKey):
    url = f"https://apis.openapi.sk.com/transit/routes?lang=0&format=json&appKey={appKey}"
    headers = {"Content-Type": "application/json"}
    body = {
        "startX": str(startX),
        "startY": str(startY),
        "endX": str(endX),
        "endY": str(endY),
        "count": 1
    }
    response = requests.post(url, headers=headers, json=body)
    if response.status_code == 200:
        try:
            path = response.json()["metaData"]["plan"]["itineraries"][0]
            return ("대중교통", path["totalDistance"], path["totalTime"])
        except:
            return ("대중교통", None, None)
    return ("대중교통", None, None)

def recommend_best_route(startX, startY, endX, endY, appKey):
    results = []
    for func in [get_walk_route, get_car_route, get_transit_route]:
        method, distance, time = func(startX, startY, endX, endY, appKey)
        if time is not None:
            print(f"📍 {method} - 거리: {distance}m / 시간: {time // 60}분 {time % 60}초")
            results.append((method, distance, time))
        else:
            print(f"⚠️ {method} 경로를 찾을 수 없습니다.")

    if not results:
        print("❌ 세 방식 모두 실패했습니다.")
        return

    best = min(results, key=lambda x: x[2])  # 소요시간 기준 최소
    print("\n✅ 가장 빠른 이동 수단 추천:")
    print(f"👉 {best[0]} (시간: {best[2] // 60}분 {best[2] % 60}초, 거리: {best[1]}m)")



# 여기에 입력값만 바꾸세요
if __name__ == "__main__":
    # ✅ 출발지 경도, 위도 (예: 위랩스)
    startX = 127.052427705027
    startY = 37.5125058180695

    # ✅ 도착지 경도, 위도 (예: SK쉴더스)
    endX = 127.098937173285
    endY = 37.4044515123315

    # ✅ 본인의 Tmap 발급 앱키
    appKey = "KXgnElYbnf9P4rPyZzanN91cHMyScabF1ZY4ifZR"

    recommend_best_route(startX, startY, endX, endY, appKey)
```
